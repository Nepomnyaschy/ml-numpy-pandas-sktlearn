## Задания
1. Довольно часто в машинном обучении исходные данные представлены разрозненным набором таблиц, с которым методы машинного обучения работать не могут. Необходимо создавать единый набор данных. Скачайте данные с конкурса [Avito Context Ad Clicks](https://www.kaggle.com/c/avito-context-ad-clicks) и объедините их в одну таблицу средствами Pandas. *Если возникнут проблемы с памятью, допускается выполнить задание лишь для части данных.*

2. Бывают случаи, когда файл с данными, хранящийся на диске, превосходит размеры доступной оперативной памяти. Далеко не все реализации алгоритмов машинного обучения способны работать с данными во внешней памяти. 
  1. Реализуйте способ загрузить данные из some_huge_file.csv (19Gb) в оперативную память средствами Pandas. Файл содержит таблицу где:
    - первый столбец - id,
    - второй - некоторые дата и время (строка, неоходимо перевести в datetime),
    - следующие 100 столбцов - категориальные признаки с 15 различными значениями (строки длиной 1000 символов),
    - оставшиеся столбцы содержат целочисленные значения от 0 до 255.
    
  2. Оптимизируйте код, чтобы загруженные данные занимали не более 100Mb.
    
  3. Время чтения файлов также играет важную роль. Оптимизируйте код и сохраните данные таким образом, чтобы время чтения было как можно меньше.

3. Решающие деревья (с коротыми вы познакомитесь в следующей лабораторной) рашают задачу классификации или регрессии за счет разбиения признакового пространства. Эффективность данного алгоритма зависит от того, какие разбиения дерево сможет построить, а какие нет. Существует способ помочь дереву строить такие разбиения. Необходимо сгруппировать данные по некоторому категориальному признаку, затем добавить к данным признаки, которые являются средним, максимальным, минимальным, и т.д. (проявите фантазию) для выбранного подмножества вещественных признаков на данных группах. Сгенерируйте данные, содержащие такой категориальный и вещественные признаки и реализуйте данный подход.

4. Встречаются задачи так называемого multiple-instance learning, когда, например, классифицировать необходимо не каждый объект отдельно, а совокупность объектов. Примером такой задачи может служить идентификация пользователя по трафику этого пользователя. Пользователь генерирует множество потоков трафика, при этом почти невозможно по единичному потоку идентифицировать пользователя, но уже по потокам, собранным за некоторый промежуток, эта задача вполне решается (в известных пределах конечно же).
  1. Сгенерируйте данные согласно следующим требованиям:
    - категориальный признак - id  пользователя кому принадлежит поток (кол-во пользователей выберите произвольно),
    - признак, содержащий дату и время потока,
    - категориальный признак - доменное имя (2го уровня) из потока (список допустимых доменов задайте самостоятельно),
    - несколько вещественных признаков - некоторые числовые характеристики потока (не обязательно, чтобы они имели физический смысл),
  2. Нарежьте сгенерированные данные на непересекающиеся сектора. Сектором назовем все потоки конкретного пользователя, взятые в некоторый интервал времени (например, 30 минут).   
  3. Для каждого сектора посчитайте частотные характеристики потоков (как в предыдущем задании).
  4. Для каждого сектора получите список посещенных доменов.
  5. Результатом дожна стать таблица, где строки - это сектор, а столбцы - это признаки из п.3 и п.4.
